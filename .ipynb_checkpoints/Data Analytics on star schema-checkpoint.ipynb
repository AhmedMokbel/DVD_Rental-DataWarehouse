# Perform some simple data analysis

### In which countries do Rent A Film have a presence in and what is the customer base in each country? What are the total sales in each country?

```python
%%sql
select c.country , count(Distinct c.customer_id) as customer_base ,sum(s.sales_amount) as total_revenue 
From star_schema.dimCustomer c 
join star_schema.FactSales s ON(c.customer_key=s.customer_key) 
Group by c.country 
order by customer_base desc  ; 

```

## Who are the top 5 customers per total sales and can we get their detail just in case Rent A Film wants to reward them?

```python
%%sql
select c.customer_id , concat(c.first_name,' ',c.last_name) as full_name , sum(F.sales_amount) as total_revenue
from star_schema.dimCustomer c
join star_schema.FactSales F ON (c.customer_key=F.customer_key) 
Group by 1,2
order by total_revenue desc
limit 5;
```

### which categories of films is most profit

```python
%%sql 
select M.category ,sum(F.sales_amount) as total_revenue 
from star_schema.dimMovie M
join star_schema.FactSales F ON (M.movie_key=F.movie_key) 
Group by M.category 
order by total_revenue desc;
```
